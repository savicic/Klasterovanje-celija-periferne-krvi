```python
import pandas as pd
import Klasterovanje as kl
```

```python
gsm_2 = pd.read_csv("GSM3892572_r_t.csv")
gsm_2
```

```python
gsm_3 = pd.read_csv("GSM3892573_r_t.csv")
gsm_3
```

```python
gsm_4 = pd.read_csv("GSM3892574_r_t.csv")
gsm_4
```

```python
gsm_5 = pd.read_csv("GSM3892575_r_t.csv")
gsm_5
```

```python
data = gsm_2.append(gsm_3).append(gsm_4).append(gsm_5)
```

```python
data
```

```python
data = data.reset_index()
```

```python
data
```

```python
features = list(data.columns)[1:]
data = data[features]
data
```

```python
file_name = "grupa_2"
```

```python
# PCA i TSNE
data_pca = kl.pca(data)
data_pca_tsne = kl.tsne(data_pca, file_name)
```

```python
# Procena epsilon knee metodom
kl.knee(data, file_name)
```

```python
labels = pd.DataFrame(data_pca_tsne, columns=['tsne_1', 'tsne_2', 'DBScan', 'KMeans', 'Agglomerative', 'Spectral'])
```

```python
# DBScan
dbscan_labels = kl.dbscan(data, data_pca_tsne, 0.1, file_name)
labels['DBScan'] = dbscan_labels
labels.to_csv(file_name + "_labels.csv", index=False)
```

```python
# KMeans
kmeans_labels = kl.kmeans(data, data_pca_tsne, file_name)
labels['KMeans'] = kmeans_labels
labels.to_csv(file_name + "_labels.csv", index=False)
```

```python
# Agglomerative
agg_labels = kl.agglomerative(data, data_pca_tsne, file_name)
labels['Agglomerative'] = agg_labels
labels.to_csv(file_name + "_labels.csv", index=False)
```

```python
data = gsm_2.append(gsm_3).append(gsm_4).append(gsm_5)
data = data.reset_index()
index = list(data['Index'])
labels = labels[['tsne_1', 'tsne_2', 'DBScan', 'KMeans', 'Agglomerative']]
labels_ = pd.DataFrame(index, columns=['Index'])
labels_[['tsne_1', 'tsne_2', 'DBScan', 'KMeans', 'Agglomerative']] = labels
```

```python
labels_.to_csv("grupa_2_labels.csv", index=False)
data.to_csv("grupa_2.csv", index=False)
```
