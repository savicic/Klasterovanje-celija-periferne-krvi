```python
import numpy as np
import pandas as pd
```

```python
valid_genes = pd.read_csv("Valid_genes.csv")
valid_genes.head()
```

```python
valid_genes=list(valid_genes['Index'])
valid_genes.sort()
len(valid_genes)
```

```python
files =['GSM3892570.csv', 'GSM3892571.csv', 'GSM3892572.csv', 'GSM3892573.csv', 'GSM3892574.csv', 'GSM3892575.csv', 'GSM3892576.csv']
columns=['Index', 'GSM3892570', 'GSM3892571', 'GSM3892572', 'GSM3892573', 'GSM3892574', 'GSM3892575', 'GSM3892576', 'num_of_cells']
```

```python
genes_info = pd.DataFrame(columns=columns)
genes_info['Index'] = valid_genes
genes_info.head()
```

```python
num_of_cells = np.zeros(len(valid_genes))

for i in range(len(files)):
    df = pd.read_csv(files[i])
    df_genes = list(df['Index'])
    sum_values=[]
    for gene in valid_genes:
        ind = df_genes.index(gene)
        df_gene = list(df.loc[ind, :])
        sum_=len(df_gene) - df_gene.count(0) -1
        sum_values.append(sum_)
        num_of_cells[valid_genes.index(gene)] += len(df_gene[1:]) 
    genes_info[columns[i+1]]=sum_values
    
    
    
```
