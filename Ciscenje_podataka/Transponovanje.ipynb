```python
import pandas as pd
import numpy as np
```

```python
chl = pd.read_csv("common_human_list.csv")
```

```python
chl.head(100)
```

```python
ensembl_GRCh38 = list(chl['Ensembl_GRCh38.p12_rel94'])
ensembl_GRCh38
```

```python
chl_genes = list(chl['ENSG_ID'])
chl_genes
```

```python
files = ['GSM3892570_.csv', 'GSM3892571_.csv', 'GSM3892572_.csv', 'GSM3892573_.csv', 'GSM3892574_.csv', 'GSM3892575_.csv', 'GSM3892576_.csv']

for file in files:
    data = pd.read_csv(file)
    genes = list(data['Index'])
    new_names=[]
    for gene in genes:
        index = chl_genes.index(gene)
        ensemble = ensembl_GRCh38[index]
        j = ensemble.rfind('.')
        if j == -1:
            new_name = gene[:4] + gene[9:] + ensemble
        else:
            new_name = gene[:4] + gene[9:] + ensemble[:j]
        new_names.append(new_name)
    #print('NOVA LISTA', new_names)
    data['Index'] = new_names 
    new_file = file[:file.rfind('.')] + 'r' + '.csv'
    data.to_csv(new_file, index = False)
```

```python
files = ['GSM3892570_r.csv', 'GSM3892571_r.csv', 'GSM3892572_r.csv', 'GSM3892573_r.csv', 'GSM3892574_r.csv', 'GSM3892575_r.csv', 'GSM3892576_r.csv']
for file in files:
    
    df = pd.read_csv(file)
    df_t = df.transpose()
    df_t.columns = df_t[0:1].values.tolist()[0]
    df_t = df_t.drop(df_t.index[0])
    new_file_name = file[:-4] + "_t.csv"
    df_t.to_csv(new_file_name)
    df = pd.read_csv(new_file_name)
    c = list(df.columns)
    c[0] = 'Cell'
    df.columns = c
    df.to_csv(new_file_name, index = False)
    
```
